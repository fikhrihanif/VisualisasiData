```python
import pandas as pd
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Age', y='Sleep Duration', data=df, errorbar=None)

plt.title('Sleep Duration by Age')
plt.xlabel('Age')
plt.ylabel('Sleep Duration (hours)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(y='Sleep Duration', x='Physical Activity Level', data=df, errorbar=None)
plt.title('Physical Activity Level by Sleep Duration')
plt.ylabel('Sleep Duration (hours)')
plt.xlabel('Physical Activity Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df = df[df['AgeGroup'] != '20-25']

category = df['AgeGroup'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category, labels=category.index, autopct='%1.1f%%', pctdistance=0.8, startangle=90, labeldistance=None)
plt.title('Age Group Category')
plt.axis('equal')

plt.legend(labels=category.index, title="Age Group", loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
plt.figure(figsize=(10, 5))

sns.scatterplot(x='Quality of Sleep', y='Stress Level', data=df)
plt.title('Sleep Quality vs Stress Level')
plt.xlabel('Sleep Quality')
plt.ylabel('Stress Level')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.hist(df['Sleep Duration'], bins=10, edgecolor='black')
plt.title('Distribution of Sleep Duration')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix = df.select_dtypes(include=['number']).corr()

plt.figure(figsize=(10, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Sleep Health and Lifestyle Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='AgeGroup', y='Sleep Duration', data=df)
plt.title('Sleep Duration Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Sleep Duration (hours)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Physical Activity Level', y='Sleep Duration', hue='Gender', data=df)
plt.title('Physical Activity Level vs. Sleep Duration by Gender')
plt.xlabel('Physical Activity Level')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```
